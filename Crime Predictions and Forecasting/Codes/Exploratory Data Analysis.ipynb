## **Summary Observations**
### **Correlations:**
- High correlated features are mainly based on dates and locations

### **Arrest & domestic cases**
- Ratio of crimes records that are non-arrestable and non-domestic is fairly high 

### **Crimes by date & time**
- Crimes rates is decreasing every year
- Crimes rates tend to increase mid year (summer)
- High number of crimes reported in the first day of the month and low on the last day of the month
- High number of crimes seen on weekends compared to weekdays
- Low crime rate in the early hours of the day but an increasing trend from lunch towards midnight

### **Crimes by location**
In order of highest to lowest:
- District 11: Harrison 
- District 8: Chicago Lawn
- District 6: Gresham
- District 4: South Chicago
- District 7: Englewood
- District 25: Grand Central
- District 3: Grand Crossing
- District 12: Near West
- District 9: Deering
- District 18: Near North

### **Types of crimes**
- Mainly theft, battery, criminal damage, narcotics, assault
- Crime type differs at different time of the day - generally theft, battery and criminal damage is commonly seen throughout the hours
- Out of the top 3 crimes, locations such as apartment, residence and streets are commomn place of crime

### **Time-series**
- 3 phases of crime rates recorded
- First downwards trend, Second: constant trend, Third - downwards trend (exponential) 

### **Mapping**
- Distribution of crimes are everywhere within the area

## Time Series EDA

```python
# Save checkpoint
ts_crimes = crimes1.copy()

# Removing time from 'date' (changes the values to object type)
ts_crimes['date'] = ts_crimes['date'].dt.date

# Change 'date' dtype to datetime
ts_crimes['date'] = pd.to_datetime(ts_crimes['date'])
```

```python
ts_crimes.info()
```

```python
# Reference code tested earlier
# crime_count = pd.DataFrame(s.groupby('primary_type').size().sort_values(ascending=False).rename('counts').reset_index())

# Transform to time series data
ts_crimes1 = ts_crimes.groupby('date').count()[['primary_type']]

# Transform to timeseries data (month)
ts_crimes_mth = ts_crimes1.resample('M').sum()

# Filter data based on 'arrest' and transform to timseries data
arrest = ts_crimes[ts_crimes['arrest'] == True]
arrest = arrest.groupby('date').count()[['primary_type']]

# Filter data based on 'non_arrest' and transform to timseries data
non_arrest = ts_crimes[ts_crimes['arrest'] != True]
non_arrest = non_arrest.groupby('date').count()[['primary_type']]

# Filter data based on 'domestic' and transform to timseries data
domestic = ts_crimes[ts_crimes['domestic'] == True]
domestic = domestic.groupby('date').count()[['primary_type']]

# Filter data based on 'non_domestic' and transform to timseries data
non_domestic = ts_crimes[ts_crimes['domestic'] != True]
non_domestic = non_domestic.groupby('date').count()[['primary_type']]
```

## Mapping Crime Locations


### Theft Crime Mapping

```python
theft_crime = crimes1[((crimes1['year'] == 2020) & (crimes1['primary_type'] == 'THEFT'))]
```

```python
theft_crime = theft_crime.loc[:, ['latitude', 'longitude', 'location_description', 'arrest', 'primary_type']]
```

```python
theft_crime.head()
```

```python
chicago_map = folium.Map(location=[41.864073,-87.706819],
                        zoom_start=11,
                        tiles="Stamen Toner")
```

```python
from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster(
    name='clustered icons',
    overlay=True,
    control=True,
    icon_create_function=None
)

# Assign a location based on each row lat and long
# for i in range(len(loc_2020)):
for i in range(len(theft_crime)):
  lat = theft_crime.iloc[i][0]
  long = theft_crime.iloc[i][1]
  popup_text = """Community Index : {}<br>
              Primary Type : {}<br>
              Arrest : {}<br>
              Location Description : {}<br>"""
  popup_text = popup_text.format(theft_crime.index[i],
                            theft_crime.iloc[i][-1],
                            theft_crime.iloc[i][-2],
                            theft_crime.iloc[i][-3]                             
                            )

  marker = folium.Marker(location = [lat, long], popup= popup_text,  icon=folium.Icon(color='blue', icon='info-sign'))
  marker_cluster.add_child(marker)

marker_cluster.add_to(chicago_map)
folium.LayerControl().add_to(chicago_map)

# save map to html file
chicago_map.save('/content/drive/MyDrive/Colab Notebooks/theft_mapping.html')
```
