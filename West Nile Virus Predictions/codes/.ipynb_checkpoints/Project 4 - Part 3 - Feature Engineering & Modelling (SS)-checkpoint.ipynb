```python
# Look at the heatmap to understand correlation within the fetaures 
def plot_heatmap(dataset, size_x=8, size_y=8, title="", title_font_size=12):
    plt.figure(figsize=(size_x,size_y))

    mask = np.zeros_like(dataset.corr(), dtype=bool)
    mask[np.triu_indices_from(mask)] = True

    hm = sns.heatmap(dataset.corr(), mask=mask, cmap='coolwarm',vmin=-1, vmax=1,
                     square=True, linewidth=1,annot=True,annot_kws={'size':'11'},
                     cbar_kws={"shrink": 0.8}, fmt='.1g',center=0)

    hm.set_xticklabels(hm.xaxis.get_ticklabels(), fontsize=14)
    hm.set_yticklabels(hm.yaxis.get_ticklabels(), fontsize=14)

    hm.set_title(title, fontsize=title_font_size);
    
    return hm

plot_heatmap(merge_df_FE, size_x=22, size_y=22, title='Correlation Heatmap for Merged Data', title_font_size=26);
```

From the correlation heatmap we found that there are many temperature features which are correlated to each other. Here we try to drop the unwanted temperature features which are highly correlated to each other. So we drop Tmin, Tmax, heat and cool as essentially they indicate the general temperature of the day. 

```python
# Drop correlated temperature columns from the  train and test dataset
merge_df_FE = merge_df_FE.drop(columns=['tmin', 'tmax', 'heat', 'cool'])
merge_test_df = merge_test_df.drop(columns=['tmin', 'tmax', 'heat', 'cool'])
```

## Finding Relative Humidity

```python
# Function to calculate relative humidity
def relative_humidity(tavg,dewpoint):
    #converted temperature to celcius
    tavg_celcius = (5.0/9.0)*(tavg-32.0)
    dewpoint_celcius = (5.0/9.0)*(dewpoint-32.0)
    
    # Calculated vapour pressure and actual pressure
    saturation_vapour_pressure = 6.11*10.0**(7.5*tavg_celcius/(237.7+tavg_celcius))
    actual_pressure = 6.11*10.0**(7.5*dewpoint_celcius/(237.7+dewpoint_celcius))
    
    #return relative humidity
    return round(((actual_pressure/saturation_vapour_pressure)*100))
```

```python
# Calculate Relative Humidity based on dewpoint temperature on train and test datasets
merge_df_FE['relative_humidity'] = relative_humidity(merge_df_FE['tavg'],merge_df_FE['dewpoint']).astype(int)
merge_test_df['relative_humidity'] = relative_humidity(merge_test_df['tavg'],merge_test_df['dewpoint']).astype(int)
```

## One-hot encoding of related non-numeric variables

We perform one-hot encoding the following variables

1. species
2. codesum

```python
# One hot encoding for species for merged train and test datasets
merge_df_FE = pd.get_dummies(merge_df_FE, columns=['species'], drop_first=True)
merge_test_df = pd.get_dummies(merge_test_df, columns=['species'], drop_first=True)
```

```python
merge_df_FE
```

```python
# One hot encoding for codesum
# Split the codesum column into multiple columns by comparing to all valid codes

valid_codes = ['+FC','FC', 'TS', 'GR', 'RA', 'DZ', 'SN', 'SG', 'GS', 'PL',
         'IC', 'FG+', 'FG', 'BR', 'UP', 'HZ', 'FU', 'VA', 'DU', 'DS',
         'PO', 'SA', 'SS', 'PY', 'SQ', 'DR', 'SH', 'FZ', 'MI', 'PR',
         'BC', 'BL', 'VC']

def conver_codesum_to_col(dataset, valid_code_values):
    for col in valid_code_values:
        dataset[col] = dataset['codesum'].apply(lambda x: 1 if col in x else 0)
        if dataset[col].sum() == 0:
            dataset.drop(columns=[col], inplace=True)
    dataset.drop(columns='codesum', inplace=True)
  
```

```python
# Convert codesum to one-hit encoded columns for train dataset
conver_codesum_to_col(merge_df_FE, valid_codes)
merge_df_FE  
```

```python
# Convert codesum to one-hit encoded columns for train dataset
conver_codesum_to_col(merge_test_df, valid_codes)
merge_test_df    
```

## Model Preparation

```python
# List of models that will be used
# Instiantiate models
models = {'lr': LogisticRegression(max_iter=5_000, random_state=123, solver='liblinear'),
          'rf': RandomForestClassifier(random_state=123),
          'gb': GradientBoostingClassifier(random_state=123),
          'dt': DecisionTreeClassifier(random_state=123),
          'et': ExtraTreesClassifier(random_state=123),
          'ada': AdaBoostClassifier(random_state=123)
        }
```

```python
def plot_roc_auc_curve(model, X_test_values, y_test_values):
    
    y_pred_proba = model.predict_proba(X_test_values)[:,1]
    fpr, tpr, _ = roc_curve(y_test_values,  y_pred_proba)
    auc = roc_auc_score(y_test_values, y_pred_proba)
    
    plt.plot(fpr,tpr,label="data, auc=" + str(round(auc,4)))
    plt.title("ROC Curve")
    plt.xlabel('Specificity')
    plt.ylabel('Sensitivity')
    plt.plot([0, 1], [0, 1], label='baseline', linestyle='--')
    plt.legend(loc=4)
    plt.show()
    
    return auc
```

```python
def analyse_model_performance(model, X_train_values, y_train_values, X_test_values, y_test_values):
    
    
    # Get predictions
    preds = model.predict(X_test_values)

    # Save confusion matrix values
    tn, fp, fn, tp = confusion_matrix(y_test_values, preds).ravel()

    dict_scores = {'model': str(model.estimator.steps),
                'cross_val_score': model.best_score_, 
                'train_auc_score': model.score(X_train_values, y_train_values),
                'test_auc_score': model.score(X_test_values, y_test_values),
                'train_accuracy_score': accuracy_score(y_train_values, model.predict(X_train_values)),
                'test_accuracy_score': accuracy_score(y_test_values, model.predict(X_test_values)),               
                'sensitivity': tp / (tp + fn),
                'specificity': tn / (tn + fp)
              }

    print(f"****Best Params****: \n{model.best_params_}\n")
    print(f"****Cross Val Score****: \n{dict_scores['cross_val_score']}\n")
    print(f"****Train AUC Score****: \n{dict_scores['train_auc_score']}\n")
    print(f"****Test AUC Score ****: \n{dict_scores['test_auc_score']}\n")
    print(f"****Train Accuracy Score****: \n{dict_scores['train_accuracy_score']}\n")
    print(f"****Test Accuracy Score ****: \n{dict_scores['test_accuracy_score']}\n")
    print(f"****Sensitivity****: \n{dict_scores['sensitivity']}\n")    
    print(f"****Specificity****: \n{dict_scores['specificity']}\n")

    plt.figure(figsize=(9,9));
    cf = plot_confusion_matrix(model, X_test_values, y_test_values, display_labels=['WnV Absent', 'WnV Present'], cmap='Blues', values_format='d');
    plt.show();
    
    plt.figure(figsize=(6,6));
    auc = plot_roc_auc_curve(model, X_test_values, y_test_values);
    plt.show();
    
    return dict_scores
```

```python
def fit_model(model, pipe_params={}, gridsearch=False, X_train_loc=X_train_sc, X_test_loc=X_test_sc, y_train_loc=y_train, y_test_loc=y_test):
    pipe = Pipeline([(model, models[model])])

    params = pipe_params

    scoring={'AUC': 'roc_auc'}
    
    if (gridsearch == True):
        gs = GridSearchCV(pipe, params, scoring=scoring, cv=3, refit='AUC', verbose=2)
        gs.fit(X_train_loc, y_train_loc)
        pipe = gs
    else:
        pipe.fit(X_train_loc, y_train_loc)

    score_dict = {}
    score_dict = analyse_model_performance(pipe, X_train_loc, y_train_loc, X_test_loc, y_test_loc)
    score_dict['model'] = model
    
    return pipe, score_dict
```
